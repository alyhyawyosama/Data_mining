```python
import pandas as pd
```

```python
data = pd.read_csv ("./train-data.csv")
```

## Drop Unwanted Columns


```python
data.drop(["Unnamed: 3","Year"],axis=1,inplace=True)
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
y = pd.DataFrame(data, columns = ["Price"])
x = pd.DataFrame(data, columns = ['Kilometers_Driven'])

```

```python
data.isnull().sum()
```

```python
plt.scatter(x, y, alpha = 0.3)


plt.ylabel('Price')
plt.xlabel("Kilometers_Driven")

# y.max()
# x.max()
plt.ylim(0,90000)
plt.xlim(0, 250000)
```

```python
import sklearn
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
reg = LinearRegression()
```

```python
reg.fit(x, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
x_pred_test = reg.predict(x)     # predicted value of y_test
x_pred_test
```

```python
plt.scatter(x, y, alpha = 0.3)

plt.xlabel("Kilometers_Driven")
plt.ylabel("Price")


plt.plot(x, reg.predict(x), color='red')
# plt.plot(x, (y), color='red')
plt.box(False)
plt.show()     
```
